## a)

We are asked to show that $B_{i,n}(t) \ge 1$ for all $t \in [0,1]$. $B_{i,n}(t)$ is defined as the following

$$
B_{i,n}(t) = \binom{n}{i}t^i(1-t)^{n-i}, \qquad i = 0, \dots , n
$$

To prove that $B_{i,n}(t) \ge 0$ we devide the expression into 3 factors, and show that all of them must be greater then zero. Ergo, the final expression must be greater the 0. 
We start by showing that $\binom{n}{i} \ge 0$.

$$
\binom{n}{i} = \frac{n!}{i!(n-i)!}
$$

Since $i=0,\dots, n$, we know that $n-i$ has to be a natural number bewteen 0 and n, which means $n-i = N\in[0,n]$. The factorial of a positive integer is a positive. A fraction of only positive integers has has to be greater or equal to 0, $\binom{n}{i} \ge 0$.

Since $t \in [0,1]$, we know that $t^i \ge 0$ for $ i= 0,1, \dots, n$

Using the same reasoning again, we know that $(1-t) \in [0,1]$ since $t \in [0,1]$. This means that $(1-t)^{n-i} \ge 0$ for $ i= 0,1, \dots, n$.

Since $B_{i,n}(t)$ is the product of 3 values greater or equal to 0, we know that $B_{i,n}(t)$ has to be greater or equal to 0, which was to be shown.

$$
B_{i,n}(t) \ge 0
$$



## b)

We are asked to show that the following statment is true
$$
B_{i,n}(t) = (1-t)B_{i,n-1}(t) + tB_{i-1,n-1}(t) \qquad t\in[0,1]
$$

for $0<i<n$. 

We start by defining $B_{i,n-1}(t)$ and $B_{i-1,n-1}(t)$

$$
B_{i,n-1}(t) = \frac{(n-1)!}{i!(n-i-1)!}t^i(1-t)^{n-i-1}
$$

$$
B_{i-1,n-1}(t) = \frac{(n-1)!}{(i-1)!((n-1)-(i-1))!}t^{i-1}(1-t)^{(n-1)-(i-1)}
$$

$$
 = \frac{(n-1)!}{(i-1)!(n-i)!}t^{i-1}(1-t)^{n-i}
$$

We combine this with the original expression 

$$
B_{i,n}(t) = (1-t)\frac{(n-1)!}{i!(n-i-1)!}t^i(1-t)^{n-i-1}\\  + t\frac{(n-1)!}{(i-1)!((n-1)-(i-1))!}t^{i-1}(1-t)^{(n-1)-(i-1)}

$$
=\frac{(n-1)!}{i!(n-i-1)!}t^i(1-t)^{n-i}  + \frac{(n-1)!}{(i-1)!((n-1)-(i-1))!}t^{i}(1-t)^{(n-1)-(i-1)}
$$


$$
=t^i (1-t)^{n-i}[\frac{(n-1)!}{i!(n-i-1)!} + \frac{(n-1)!}{(i-1)!(n-i)!}]
$$

We do some maniplation of the expression inside the brackets.

$$
\frac{(n-1)!}{i!(n-i-1)!} \cdot \frac{n-i}{n-i} + \frac{(n-1)!}{(i-1)!(n-i)!} \cdot \frac{i}{i}
$$

$$
\frac{(n-1)!}{i!(n-i)!} \cdot (n-i) + \frac{(n-1)!}{i!(n-i)!} \cdot i
$$


$$
= \frac{(n-1)!}{i!(n-i)!}(n-i+i)= \frac{(n-1)!}{i!(n-i)!}\cdot n=\frac{n!}{i!(n-i)!}
$$

We now return to the original expression.

$$
B_{i,n}(t) =t^i (1-t)^{n-i}\frac{n!}{i!(n-i)!}
$$

$$
B_{i,n}(t) =\binom{n}{i}t^i (1-t)^{n-i}
$$

This means the recursive formula has to be true for $0 < i < n$.For $i=0$ and $i=n$ we get a special case.

If we plug $i=0$ in to the formula for Bernstein polynomials we get the following

$$
B_{0,n}(t) = \binom{n}{0}t^0(1-t)^{n-0}
$$

$$
= 1 \cdot 1 \cdot (1-t)^n = (1-t)^n
$$

If we plug $i=n$ in to the formula for Bernstein polynomials we get the following

$$
B_{i,n}(t) = \binom{n}{n}t^n(1-t)^{n-n}, \qquad i = 0, \dots , n
$$
$$
= 1 \cdot t^n \cdot (1-t)^0 = t^n
$$

This gives the following result

$$
B_{i,n}(t) = 
\begin{cases}
(1-t)^n, & \text{if } i = 0 \\
t^n, & \text{if } i = n \\
(1-t)B_{i,n-1}(t) + tB_{i-1,n-1}(t), & \text{if } 0<i<n \\
\end{cases}
$$

## c)

We are asked to show that $B_{i,n}(t)$ form a partition of unity, which can be written as the following. 

$$
\sum^n_{i=0}B_{i,n}(t)=1, \qquad \text{for} \space t\in [0,1]
$$

We will do this by induction. We assume that $\sum^n_{i=0}B_{i,n}(t)=1$, and want to show that this necessarily indicates that $\sum^{n+1}_{i=0}B_{i,n+1}(t)=1$. We will also use the expression for $B_{i,n}(t)$ we found in $\mathbf{b)}$.

$$
\sum^{n+1}_{i=0}B_{i,n+1}(t)=\sum^{n+1}_{i=0}(1-t)B_{i,n}(t)+t\cdot B_{i-1,n}(t)
$$

$$
=(1-t)\cdot \sum^{n+1}_{i=0}B_{i,n}(t)+t\cdot \sum^{n+1}_{i=0}B_{i-1,n}(t)
$$

$$
=(1-t)\cdot [\sum^{n}_{i=0}B_{i,n}(t)+B_{n+1,n}(t)] +t\cdot[ \sum^{n+1}_{i=1}B_{i-1,n}(t) +B_{0-1,n}(t) ]
$$

We assume that $B_{n+1,n}(t) = B_{0-1,n}(t) =0 $. 
We observe that $\sum^{n+1}_{i=1}B_{i-1,n}(t) = \sum^{n+1}_{i=0}B_{i,n}(t) $. From our original assumption we can repace $\sum^{n+1}_{i=0}B_{i,n}(t)$ with $1$. If we plug all this into the original expression we get the following

$$
=(1-t)\cdot [1+0] +t\cdot[1+0] = 1-t+t = 1
$$

All we have left to check is that the claim $\sum^n_{i=0}B_{i,n}(t)=1$ holds for $n=0$. 

$$
\sum^0_{i=0}B_{i,n}(t)= B_{0,0}(t) =  \binom{0}{0} t^0(1-t)^0=1
$$

We have shown by induction that $\sum^n_{i=0}B_{i,n}(t)=1$ for $t\in[0,1]$.

We are asked to show that the derivative of $B_{i,n}(t)$ is given by the following expression

$$
B'_{i,n}(t)=n[B_{i-1,n-1}(t)-B_{i,n-1}(t)]
$$

We will show this by simply differentiating the expression for $B_{i,n}(t)$

$$
B_{i,n}(t)=\binom{i}{n}t^i(1-t)^{n-i}, \qquad i=0,\dots,n
$$

$$
\frac{d B_{i,n}(t)}{dt} = \binom{i}{n}[i\cdot i^{i-1}(1-t)^{n-i} + t^i(1-t)^{n-i-1}\cdot(n-i)\cdot(-1)]
$$

We start by evaluating the first element
$$
\binom{i}{n}\cdot i\cdot i^{i-1}(1-t)^{n-i} 
$$

$$
= \frac{n!\cdot i}{i!(n-i)!} \cdot t^{i-1}(1-t)^{n-i} 
$$

$$
= \frac{(n-1)!\cdot n}{((i-1)!((n-1)-(i-1))!} \cdot t^{i-1}(1-t)^{(n-1)-(i-1)}

$$
= n \cdot B_{i-1,n-1}(t)
$$

We now look at the second element 

$$
-\binom{i}{n}t^i(1-t)^{n-i-1}\cdot(n-i) 
$$

$$
=- \frac{n!\cdot (n-i)}{i!(n-i)!}t^i(1-t)^{n-i-1}

$$

$$

When combining the two expressions again, we get the following result

$$

## e)

We are asked to impImplementa function $\mathbf{Bernstein(i,n,t)}$, and plot the function $B_{i,n}, \space i=0,\dots,n$ for $n=4,5,15$. Implementation of the function can be seen in the cell belove, and the plots in the block underneath. 

```python
def Bernstein(i,n,t):
    """
    Evaluates B_{i,n} at t
    
    input: 
    i - i value, n - upper limit, t - time
    
    return:
    B_{i,n}(t)
    """
    value = math.factorial(n)/(math.factorial(i)*math.factorial(n-i))*t**i*(1-t)**(n-i)
    return value
```

```python
time_values=np.linspace(0,1,1000)

n=4
Bernstein_values=[]
sum_values=np.zeros(1000)
for i in range(0,n+1):
    Bernstein_values.append(Bernstein(i,n,time_values))
    sum_values+=Bernstein_values[i]
    

plt.figure()
plt.title(r"$B_{i,n}(t)$ for $n=4$")
for j in range(n+1):
    plt.plot(time_values,Bernstein_values[j], label=f"B_{j,n}")
plt.plot(time_values, sum_values,color="black", label=r"$\Sigma_{i=0}^4 B_{i,n}$")
plt.xlabel("Time")
plt.ylabel("Weight")
plt.legend(loc="upper right")
plt.show()

n=5
Bernstein_values=[]
sum_values=np.zeros(1000)
for i in range(0,n+1):
    Bernstein_values.append(Bernstein(i,n,time_values))
    sum_values+=Bernstein_values[i]
    
plt.figure()
plt.title(r"$B_{i,n}(t)$ for $n=5$")
for j in range(n+1):
    plt.plot(time_values,Bernstein_values[j], label=f"B_{j,n}")
plt.plot(time_values, sum_values,color="black", label=r"$\Sigma_{i=0}^5 B_{i,n}$")
plt.xlabel("Time")
plt.ylabel("Weight")
plt.legend(loc="upper right")
plt.show()

n=15
Bernstein_values=[]
sum_values=np.zeros(1000)
for i in range(0,n+1):
    Bernstein_values.append(Bernstein(i,n,time_values))
    sum_values+=Bernstein_values[i]
    
plt.figure()
plt.title(r"$B_{i,n}(t)$ for $n=15$")
for j in range(n+1):
    plt.plot(time_values,Bernstein_values[j], label=f"B_{j,n}")
plt.plot(time_values, sum_values,color="black", label=r"$\Sigma_{i=0}^15 B_{i,n}$")
plt.xlabel("Time")
plt.ylabel("Weight")
plt.legend(fontsize="5", loc="upper right")
plt.show()
```

In the plots above we have plotted $B_{i,n}(t), \space i=0,\dots,n, \space t \in [0,1]$ for $n=4,5,15$. We have also plotted $\Sigma_{i=0}^n B_{i,n}(t)$ for $t \in [0,1]$, which we observe is contsantly equal to $1$ for all $t$. This confirms the result from $\mathbf{c)}$.
