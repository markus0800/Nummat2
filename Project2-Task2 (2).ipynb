### a) 
We are given the $n$th degree Bezier curve as $c(t)$:


$$
c(t) = b_n(t; p_0, \ldots, p_n) = \sum_{i=1}^n B_{i,n}(t)p_i = \binom{i}{n}t^{i}(1-t)^{n-i}p_i
$$

Using the results from $\textbf{1b)}$ and  $\textbf{1d)}$ we find the first and second derivative of $c(t)$.

$$
\begin{align*}
c^{\prime}(t) &= \frac{d}{dt} \sum_{i=0}^n B_{i,n}(t)p_i \\
&=\sum_{i=0}^n  \frac{d}{dt} \left(B_{i,n}(t)\right)p_i \\
&= \sum_{i=0}^n n \left(B_{i-1,n-1}(t) - B_{i,n-1}(t) \right) p_i \\
&= n \left(\sum_{i=0}^{n} B_{i-1,n-1}(t)p_i - \sum_{i=0}^{n} B_{i,n-1}p_i \right) \\
&= n \left[ \left(\sum_{i=0}^{n-1} B_{i,n-1}(t)p_{i+1}\right) + B_{-1,n-1}(t)p_0 - \left(\sum_{i=0}^{n-1} B_{i,n-1}p_i\right) - B_{n,n-1}p_n\right] \\ 
&= n \left[ \left(\sum_{i=0}^{n-1} B_{i,n-1}(t)p_{i+1}\right) + 0 - \left(\sum_{i=0}^{n-1} B_{i,n-1}p_i\right) - 0\right] \\ 
&= n \left[ \sum_{i=0}^{n-1} B_{i,n-1}(t)\left(p_{i+1} - p_i\right)\right] 
\end{align*}
$$

and for the second order derivative we get

$$
\begin{align*}
c^{\prime \prime}(t) &= \frac{d}{dt} c^{\prime}(t) \\
&=  \frac{d}{dt}  \sum_{i=0}^n n \left(B_{i-1,n-1}(t) - B_{i,n-1}(t) \right) p_i \\
&= n \sum_{i=0}^n \frac{d}{dt} \left(B_{i-1,n-1}(t) - B_{i,n-1}(t) \right) p_i \\
&= n \sum_{i=0}^n (n-1) \left[\left(B_{i-2,n-2}(t) - B_{i-1,n-2}(t)\right) - \left(B_{i-1,n-2}(t) - B_{i,n-2}(t)\right) \right] p_i \\
&= n(n-1) \left[  \sum_{i=0}^n B_{i-2,n-2}(t)p_i - 2\sum_{i=0}^n B_{i-1,n-2}(t)p_i + \sum_{i=0}^nB_{i,n-2}(t)p_i \right]\\
&= n(n-1) \left[ \left(\sum_{i=0}^{n-2} B_{i,n-2}(t)p_{i+2}\right) + B_{-2,n-2}(t)p_0 + B_{-1,n-2}(t)p_1 
-2 \left(\sum_{i=0}^{n-2} B_{i,n-2}(t)p_{i+1}\right) -2B_{-1,n-2}p_0 -2B_{n-1,n-2}p_n \\
+\left(\sum_{i=0}^{n-2} B_{i,n-2}(t)p_{i}\right) +  B_{n,n-2}(t)p_n + B_{n-1,n-2}(t)p_{n-1}\right] \\ 
&= n(n-1) \left[ \left(\sum_{i=0}^{n-2} B_{i,n-2}(t)p_{i+2}\right) + 0 + 0 -2 \left(\sum_{i=0}^{n-2} B_{i,n-2}(t)p_{i+1}\right) - 0 - 0 + \left(\sum_{i=0}^{n-2} B_{i,n-2}(t)p_{i}\right) + 0 + 0\right] \\
&= n(n-1) \left[ \sum_{i=0}^{n-2} B_{i,n-2}(t)\left(p_{i+2} - 2p_{i+1} + p_{i}\right) \right]
\end{align*}
$$

### b) 
Now we evaluate $c(t)$ at its end points $t= 0$ and $t = 1$. Using the results from $\textbf{1b)}$ we obtain 

$$
c(0) = \sum_{i=0}^n B_{i,n}(0)p_i = (1-0)^np_0 + 0\cdot p_1 + \ldots + 0\cdot p_{n-1}+ 0^n\cdot p_n = p_0 \\
c(1) = \sum_{i=0}^n B_{i,n}(1)p_i = (1-1)^np_0 + 0\cdot p_1 + \ldots + 0\cdot p_{n-1} + 1^n\cdot p_n  = p_n.
$$
 
We observe that for every $n$th degree Bernstein polynomial the $n$th degree Bezier curve evaluated in $t=0$ and $t=1$ is equal to the endpoints $p_0$ and $p_n$. This means that every Bezier curve allways interpolate the first and last point of our sampled points. 


### c)
We are given the following claim 

$$
\textbf{*claim:}
\begin{cases}
c^{\prime}(0) = n(p_1-p_0) \\
c^{\prime}(1) = n(p_n - p_{n-1}) \\
c^{\prime \prime}(0) = n(n-1)(p_0 - 2p_1 +p_{2}) \\
c^{\prime \prime}(1) = n(n-1)(p_n - 2p_{n-1} + p_{n-2}) \\
\end{cases}
$$

To prove that the properties above holds we use our calculated derivatives from $\textbf{2a)}$ to obtain:

$$
\begin{align*}
c'(0) &= n \left[ \sum_{i=0}^{n-1} B_{i,n-1}(0)\left(p_{i+1} - p_i\right)\right] = n \left( B_{0,n-1}(0)(p_1-p_0) + 0 + \ldots + 0 \right) = n(p_1-p_0)  \\
c'(1) &= n \left[ \sum_{i=0}^{n-1} B_{i,n-1}(1)\left(p_{i+1} - p_i\right)\right] = n \left(0 + \ldots + 0 + B_{n-1,n-1}(1)(p_n-p_{n-1}) \right) = n(p_n-p_{n-1})  \\
\end{align*}
$$

similarly for the second order derivative we obtain 

$$
\begin{align*}
c''(0) &= n(n-1) \left[ \sum_{i=0}^{n-2} B_{i,n-2}(0)\left(p_{i+2} - 2p_{i+1} + p_{i}\right) \right] 
= n(n-1)\left( B_{0,n-2}(0)(p_{2} - 2p_{1} + p_{0}) + 0 + \ldots + 0\right) = n(n-1)(p_{0} - 2p_{1} + p_{2})\\
c''(1) &= n(n-1) \left[ \sum_{i=0}^{n-2} B_{i,n-2}(1)\left(p_{i+2} - 2p_{i+1} + p_{i}\right) \right] 
= n(n-1)\left(0 + \ldots + 0 + B_{n-2,n-2}(1)(p_{n} - 2p_{n-1} + p_{n-2}) \right) = n(n-1)(p_{n} - 2p_{n-1} + p_{n-2})\\
\end{align*}
$$

wich proves the claim. 

From the results above we can spesify interpolation properties at it end points. We observe that the first order derivative at $t = 0$ is determined by the order $n$ of the curve as well as the starting point $p_0$ and the following point $p_1$. Similarly for the endpoint $t=1$, the first other derivative is determined on the last two points ($p_{n-1}$ and $p_n$) and the order $n$ of the curve. For the second order derivative (the curvature of the curve), is determined by the first and last three points to respectively $t=0$ and $t=1$, and the order $n$ of the curve. 

If we change $p_1$, we also change both the growth and the curvature of the curve, at the starting point $t=0$. I nice visual idea is to look at the first and second order derivative as respectively velocity and acceleration along the curve. When we increase the distance between the points $p_0$ and $p_1$ we also increase the speed we are traversing the curve, since we are travelling a further distance in the same time interval $\Delta t$. Similarly will the acceleration at $t=0$ be affected by where we place our point $p_1$. This is easily visiable when we look at the formula for the second order derivative at $t=0$.

When we change $p_2$ we change the starting acceleration $\vec{a}(0)$ of the curve, but the starting velocity $\vec{v}(0)$ remains the same. Take the example when we start with points $p_0 = (0,0)$ and $p_1 = (1,1)$ in $\mathbb{R}^2$, and then choose a point $p_2 = (2, 10)$. We observe that the acceleration in the $y$-direction is large, but a small acceleration in $x$-direction. Geometricy wise this means that at the startpoint $t=0$ we need to follow a curve that curves a lot upwards. 


### d) 

$\textbf{*claim:}$ $b_n(t; p_0,\ldots, p_n) = (1 − t)b_{n-1}(t; p_0,\ldots, p_{n−1}) + tb_{n−1}(t; p_1,\ldots, p_n)$

We use the recursive definition of the Berstein polynomials from $\textbf{1b)}$ and obtain:

$$
\begin{align*}
c(t) = b_n(t; p_0,\ldots, p_n) &= \sum_{i=0}^n \left((1-t)B_{i,n-1}(t) + tB_{i-1,n-1}(t)\right) p_i \\
&= \sum_{i=0}^n \left((1-t)B_{i,n-1}(t)p_i\right) + \sum_{i=0}^n \left(tB_{i-1,n-1}(t)p_i\right) \\
&=(1-t) \left(\sum_{i=0}^{n-1} \left(B_{i,n-1}(t)p_i\right) + B_{n,n-1}p_n\right) + t\left( B_{-1,n-1} p_0 + t\sum_{i=0}^{n-1} B_{i,n-1}(t)p_{i+1} \right)\\
&=(1-t) \left(\sum_{i=0}^{n-1} \left(B_{i,n-1}(t)p_i\right) + 0\right) + t\left( 0+ t\sum_{i=0}^{n-1} B_{i,n-1}(t)p_{i+1} \right)\\
&= (1-t)b_{n-1}(t; p_0,\ldots, p_{n−1}) + tb_{n−1}(t; p_1,\ldots, p_n),
\end{align*}
$$

wich proves the claim.

```python
def de_Casteljau(P,t0):
    '''
    function that implements the de Casteljau algorithm

    input:
        P: vector of points we want to make a Bezier Curve for 
        t0: t-parameter 
    output:
        Pvecs[-1,0]: the point on the Bezier curve for t0 
        Pvecs[:,0]: vector of points for every "level" k considered 
    '''
    n,d = np.shape(P) #get dimentions of P 
    Pvecs = np.zeros((n, n, d)) #create matrix of points  
    Pvecs[0,:] = P #initialize the matrix 
    for k in range(1,n): #double for-loop to calculate the Bezier curve at t0 
        for i in range(0,n-k):
            Pvecs[k,i] = (1-t0) * Pvecs[k-1,i] + t0 * Pvecs[k-1, i+1] #de Casteljau's algorithm 
    return Pvecs[-1,0], Pvecs
    

p0 = [0,0]; p1 = [-1,5]; p2 = [4,4]; p3 = [9,3]; p4 = [8,8] #points that we want to make our Bezier curve for 
P = np.array([p0, p1, p2, p3, p4]) #make it into numpy array 


T = np.linspace(0,1, 100) #our time interval 

Pvecs = np.zeros((len(T),  2))#our bezier curve for different k-levels 
Pvecs2 = np.zeros((len(T), len(P),len(P), 2))
for t in range(len(T)): #for loop that uses de Casteljaus algrorithm 
    Pvecs[t] = de_Casteljau(P, T[t])[0]


plt.figure(figsize = (10,7))    
plt.plot(P[:,0], P[:,1], 'r.') #plot the points we want to make Bezier curve for 

plt.plot(Pvecs[:,0], Pvecs[:,1],c='blue', label = f'Bezier curve')

points2 = de_Casteljau(P, 1/3)[1]
plt.plot(points2[0,:,0], points2[0,:,1], 'o', c='r', label='Control points')
plt.plot(points2[0,:,0], points2[0,:,1], 'o--', c='r', label='k=0')
plt.plot(points2[1,:4,0], points2[1,:4,1], 'x--', c='orange', label='k=1')
plt.plot(points2[2,:3,0], points2[2,:3,1], 'x--', c='green', label='k=2')
plt.plot(points2[3,:2,0], points2[3,:2,1], 'x--', c='purple', label='k=3') 
plt.plot(points2[4,:1,0], points2[4,:1,1], 's', c='blue', label=r'$c(\frac{1}{3})$')
plt.grid(True)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Bézier Curve and Intermediate Points')
plt.legend()
plt.show()
```

### f) 
For points $(p_0,p_1, \ldots, p_n)\in \mathbb{R}^d$ and $(q_0,q_1, \ldots, q_n)\in \mathbb{R}^d$ with their respective Bézier curves $c(t) = b_n(t; p_0,\ldots, p_n)$ and $d(t) = b_n(t; q_0,\ldots, q_n)$, the composite Bézier curve $s(t)$, $s:[0,2] \rightarrow \mathbb{R}^d$ is defined by: 

$$
s(t) = \begin{cases}
c(t), & 0 \leq t < 1 \\
d(t − 1), & 1\leq t \leq 2.
\end{cases}
$$

##### 1) 
Since we know that the bezier curve is a $C^{2}$ along its curve, $c$ and $d$ seperated must be $C^{2}$ functions. The intresting points are therefore the adjacent points between the curves, which is $s(1)$ or $p_{n}$ and $q_{0}$.  

##### 2)
For $s(t)$ to be an $C^0$ function it need to be continious along all of $s(t)$. We know that $c(t)$ and $d(t)$ seperated are continious. For $s(1)$ to be continious, we need the endpoint of $c$ is equal to the starting point of $d$.

$$
\begin{align}
&\lim_{t\to 1^-}s(t) = s(1) \\
\implies &\lim_{t\to 1^-}c(t)=d(0)\\
\implies & p_n=q_0
\end{align}
$$

$s(t)$ is an $C^0$ function if:

$$
\begin{cases}
p_n = q_0\\
\end{cases}
$$

##### 3)
For $s(t)$ to be an $C^1$ function it need to be continious and continious differentiable. We know that $c(t)$ and $d(t)$ seperated are continious and continious differentiable. We use the result from $1)$ to make $s(t)$ continious. For $s(1)$ to be continious differentiable, we need the derivative of the endpoint of $c$ to be equal to the derivative of the starting point of $d$.

$$
\begin{align}
&\lim_{t\to 1^-}s(t) = s(1) \\
\implies & \space{}p_n=q_0 \\
\end{align}
$$


<br>


$$
\begin{align}
&\lim_{t\to 1^-} s'(t) = s'(1) \\
\implies &\lim_{t\to 1^-} c'(t) = d'(0) \\
\text{Use result for $\textbf{c)}$}& \\
&n(p_n-p_{n-1}) = n(q_1-q_0) \\
\implies &p_n-p_{n-1} = q_1-q_0 \\
\text{Use result for $\textbf{2)}$}& \\
&2 p_n-p_{n-1} = q_1\\
\end{align}
$$

$s(t)$ is an $C^1$ function if:

$$
\begin{cases}
p_n = q_0\\
2p_n - p_{n-1} = q_1\\
\end{cases}
$$

##### 4) 
For $s(t)$ to be an $C^2$ function it need to be $C^1$ as well as twice continious differentiable. We know that $c(t)$ and $d(t)$ seperated are $C^{2}$. For $s(1)$ to be twice continious differentiable, we need the second derivative of the endpoint of $c$ to be equal to the second derivative of the starting point of $d$.

$$
\begin{align}
&\lim_{t\to 1^-}s(t) = s(1) \\
\implies & \space{}p_n=q_0 \\
\end{align}
$$

<br>

$$
\begin{align}
&\lim_{t\to 1^-} s'(t) = s'(1) \\
\implies &2 p_n-p_{n-1} = q_1 \\
\end{align}
$$

<br>

$$
\begin{align}
&\lim_{t\to 1^-} s^{\prime \prime}(t) = s^{\prime \prime}(1) \\
\implies &\lim_{t\to 1^-} c^{\prime \prime}(t)  = d^{\prime \prime}(0)\\
\implies &n(n-1)(p_n - 2p_{n-1} + p_{n-2}) =  n(n-1)(q_0 - 2q_1 +q_{2})\\
\implies &p_n - 2p_{n-1} + p_{n-2} - q_0 + 2q_1 =  +q_{2}\\
\text{Use result for $\textbf{3)}$} & \\
\implies &p_n - 2p_{n-1} + p_{n-2} - p_n + 2(2 p_n-p_{n-1})=  q_{2}\\
\implies &4 p_n - 4p_{n-1} + p_{n-2}=  q_{2}\\
\end{align}
$$

$s(t)$ is an $C^2$ function if:

$$
\begin{cases}
p_n = q_0\\
2p_n - p_{n-1} = q_1\\
4 p_n - 4p_{n-1} + p_{n-2}=  q_{2}\\
\end{cases}
$$



#### 5)

We are given 3 datapoints, $\mathbf{d_0}$,$\mathbf{d_1}$,$\mathbf{d_2}$, that must be included in $s(h)=\mathbf{d}_h$ for $h=0,1,2$. We know $s \in C^1$ and $n=3$, which means that $s(0)=p_0=d_0$,$s(1)=p_3=q_0=d_1$ and $s(2)=q_3=d_2$, as well as the following statments


$$
\begin{cases}
p_3 = q_0\\

\end{cases}
$$

The points we haven't assigned a value to are $p_1,p_2,q_1$ and $q_2$. Using the second statment, we know that $q_1$ is predetermined if we chose $p_2$. We therefore have 3 points we can chose freely, which means we have $\mathbf{3}$ degrees of freedom. 

For $n = 3$ and $s \in C^2$ we use the same reasoning to get $s(0)=p_0=d_0$,$s(1)=p_3=q_0=d_1$ and $s(2)=q_3=d_2$, aswell as 
$$
\begin{cases}
p_3 = q_0\\
2p_3 - p_{2} = q_1\\
4 p_3 - 4p_{2} + p_{1}=  q_{2}\\
\end{cases}
$$

The points we haven't assigned a value to are $p_1,p_2,q_1$ and $q_2$. Using the second and third statment above we get $q_1$ and $q_2$ are predetermined if we assign a value to $p_1$ and $p_2$. The only points we can chose freely are therefore $p_1$ and $p_2$, which means we have $\mathbf{2}$ degrees of freedom.
