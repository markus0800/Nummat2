### a)

We are asked to show that $B_{i,n}(t) \ge 1$ for all $t \in [0,1]$. $B_{i,n}(t)$ is defined as the following

$$
B_{i,n}(t) = \binom{n}{i}t^i(1-t)^{n-i}, \qquad i = 0, \dots , n
$$

To prove that $B_{i,n}(t) \ge 0$ we devide the expression into 3 factors, and show that all of them must be greater then zero. Ergo, the final expression must be greater the 0. 
We start by showing that $\binom{n}{i} \ge 0$.

$$
\binom{n}{i} = \frac{n!}{i!(n-i)!}
$$

Since $i=0,\dots, n$, we know that $n-i$ has to be a natural number bewteen 0 and n, which means $n-i = N\in[0,n]$. The factorial of a positive integer is a positive. A fraction of only positive integers has has to be greater or equal to 0, $\binom{n}{i} \ge 0$.

Since $t \in [0,1]$, we know that $t^i \ge 0$ for $ i= 0,1, \dots, n$

Using the same reasoning again, we know that $(1-t) \in [0,1]$ since $t \in [0,1]$. This means that $(1-t)^{n-i} \ge 0$ for $ i= 0,1, \dots, n$.

Since $B_{i,n}(t)$ is the product of 3 values greater or equal to 0, we know that $B_{i,n}(t)$ has to be greater or equal to 0, which was to be shown.

$$
B_{i,n}(t) \ge 0
$$



### b)

We are asked to show that the following statment is true
$$
B_{i,n}(t) = (1-t)B_{i,n-1}(t) + tB_{i-1,n-1}(t) \qquad t\in[0,1]
$$

for $0<i<n$. 

We start by defining $B_{i,n-1}(t)$ and $B_{i-1,n-1}(t)$
\begin{align*}
B_{i,n-1}(t) &= \frac{(n-1)!}{i!(n-i-1)!}t^i(1-t)^{n-i-1}\\
B_{i-1,n-1}(t) &= \frac{(n-1)!}{(i-1)!((n-1)-(i-1))!}t^{i-1}(1-t)^{(n-1)-(i-1)}\\
&= \frac{(n-1)!}{(i-1)!(n-i)!}t^{i-1}(1-t)^{n-i}\\
\end{align*}
We combine this with the original expression 

\begin{align*}
B_{i,n}(t)&= (1-t)\frac{(n-1)!}{i!(n-i-1)!}t^i(1-t)^{n-i-1}+ t\frac{(n-1)!}{(i-1)!((n-1)-(i-1))!}t^{i-1}(1-t)^{(n-1)-(i-1)} \\
&=\frac{(n-1)!}{i!(n-i-1)!}t^i(1-t)^{n-i}  + \frac{(n-1)!}{(i-1)!((n-1)-(i-1))!}t^{i}(1-t)^{(n-1)-(i-1)}\\
&=t^i (1-t)^{n-i}(\frac{(n-1)!}{i!(n-i-1)!} + \frac{(n-1)!}{(i-1)!(n-i)!})
\end{align*}
We do some maniplation of the expression inside the brackets.

\begin{align*}
&\frac{(n-1)!}{i!(n-i-1)!} \cdot \frac{n-i}{n-i} + \frac{(n-1)!}{(i-1)!(n-i)!} \cdot \frac{i}{i} \\
=&\frac{(n-1)!}{i!(n-i)!} \cdot (n-i) + \frac{(n-1)!}{i!(n-i)!} \cdot i \\
=&\frac{(n-1)!}{i!(n-i)!}(n-i+i)= \frac{(n-1)!}{i!(n-i)!}\cdot n=\frac{n!}{i!(n-i)!}\\
\end{align*}
We now return to the original expression.

\begin{align*}
&B_{i,n}(t) =t^i (1-t)^{n-i}\frac{n!}{i!(n-i)!}\\
&B_{i,n}(t) =\binom{n}{i}t^i (1-t)^{n-i}\\
\end{align*}

This means the recursive formula has to be true for $0 < i < n$.For $i=0$ and $i=n$ we get a special case.

If we plug $i=0$ in to the formula for Bernstein polynomials we get the following
\begin{align*}
B_{0,n}(t) &= \binom{n}{0}t^0(1-t)^{n-0}\\
&= 1 \cdot 1 \cdot (1-t)^n = (1-t)^n\\
\end{align*}

If we plug $i=n$ in to the formula for Bernstein polynomials we get the following

\begin{align*}
B_{n,n}(t)&= \binom{n}{n}t^n(1-t)^{n-n}\\
&= 1 \cdot t^n \cdot (1-t)^0 = t^n\\
\end{align*}

This gives the following result

$$
B_{i,n}(t) = 
\begin{cases}
(1-t)^n, & \text{if } i = 0 \\
t^n, & \text{if } i = n \\
(1-t)B_{i,n-1}(t) + tB_{i-1,n-1}(t), & \text{if } 0<i<n \\
\end{cases}
$$

### c)

We are asked to show that $B_{i,n}(t)$ form a partition of unity, which can be written as the following. 

$$
\sum^n_{i=0}B_{i,n}(t)=1, \qquad \text{for} \space t\in [0,1]
$$

We will do this by induction. We assume that $\sum^n_{i=0}B_{i,n}(t)=1$, and want to show that this necessarily indicates that $\sum^{n+1}_{i=0}B_{i,n+1}(t)=1$. We will also use the expression for $B_{i,n}(t)$ we found in $\mathbf{b)}$.

\begin{align*}
\sum^{n+1}_{i=0}B_{i,n+1}(t) &= B_{0,n+1}(t) + B_{n+1,n+1}(t) + \sum^{n}_{i=1}B_{i,n+1}(t) \\
&=B_{0,n+1}(t) + B_{n+1,n+1}(t) + \sum^{n}_{i=1}(1-t)B_{i,n}(t)+t\cdot B_{i-1,n}(t) \\
&= B_{0,n+1}(t) + B_{n+1,n+1}(t) + (1-t)\cdot \sum^{n}_{i=1}B_{i,n}(t)+t\cdot \sum^{n}_{i=1}B_{i-1,n}(t)\\
\end{align*}
 
We observe that $\sum^{n+1}_{i=1}B_{i-1,n}(t) = \sum^{n+1}_{i=0}B_{i,n}(t)$. We can  add and subtract $B_{0,n}$ to the second expression, and change the start index from $0$ to $1$. Since we have $i-1$ in the second expression we can change the sum from $\Sigma_{i=1}^n$ to $\Sigma_{i=0}^{n-1}$, and add and subtract $B_{n,n}(t)$, and again change the index to $n$ 

$$
= B_{0,n+1}(t) + B_{n+1,n+1}(t) + (1-t)\cdot (\sum^{n}_{i=0}B_{i,n}(t)- B_{0,n}) +t\cdot (\sum^{n}_{i=0}B_{i,n}(t) - B_{n,n}(t))
$$
From our original assumption we can repace $\sum^{n+1}_{i=0}B_{i,n}(t)$ with $1$. From $\mathbf{b)}$ we know that $B_{0,n+1}(t) = (1-t)^{n+1}$, $B_{0,n}(t) = (1-t)^{n}$ , $B_{n+1,n+1}(t) = t^{n+1}$, and $B_{n,n}(t) = t^{n}$. Using this, we get the following expression

\begin{align*}
\
&= (1-t)^{n+1} + t^{n+1} + (1-t)\cdot (1-(1-t)^{n}) +t\cdot (1-t^n) \\ 
&= (1-t)^{n+1} + t^{n+1} + (1-t)-(1-t)^{n+1} + t - t^{n+1} \\
&= (1-t)^{n+1}-(1-t)^{n+1} + t^{n+1}- t^{n+1} + (1-t) + t = 1 \\ 
\end{align*}

We check that the claim is correct for $n=0$.

$$
\sum^0_{i=0}B_{i,n}(t)= B_{0,0}(t) =  \binom{0}{0} t^0(1-t)^0=1
$$

We have shown by induction that $\sum^n_{i=0}B_{i,n}(t)=1$ for $t\in[0,1]$.

### e)

We are asked to impImplementa function $\mathbf{Bernstein(i,n,t)}$, and plot the function $B_{i,n}, \space i=0,\dots,n$ for $n=4,5,15$. Implementation of the function can be seen in the cell belove, and the plots in the block underneath. 

```python
def Bernstein(i,n,t):
    """
    Evaluates B_{i,n} at t
    
    input: 
    i - i value, n - upper limit, t - time
    
    return:
    B_{i,n}(t)
    """
    value = math.factorial(n)/(math.factorial(i)*math.factorial(n-i))*t**i*(1-t)**(n-i)
    return value
```

```python
time_values=np.linspace(0,1,1000)

n=4
Bernstein_values=[]
sum_values=np.zeros(1000)
for i in range(0,n+1):
    Bernstein_values.append(Bernstein(i,n,time_values))
    sum_values+=Bernstein_values[i]
    

plt.figure()
plt.title(r"$B_{i,n}(t)$ for $n=4$")
for j in range(n+1):
    plt.plot(time_values,Bernstein_values[j], label=f"B_{j,n}")
plt.plot(time_values, sum_values,color="black", label=r"$\Sigma_{i=0}^4 B_{i,n}$")
plt.xlabel("Time")
plt.ylabel("Weight")
plt.legend(loc="upper right")
plt.show()

n=5
Bernstein_values=[]
sum_values=np.zeros(1000)
for i in range(0,n+1):
    Bernstein_values.append(Bernstein(i,n,time_values))
    sum_values+=Bernstein_values[i]
    
plt.figure()
plt.title(r"$B_{i,n}(t)$ for $n=5$")
for j in range(n+1):
    plt.plot(time_values,Bernstein_values[j], label=f"B_{j,n}")
plt.plot(time_values, sum_values,color="black", label=r"$\Sigma_{i=0}^5 B_{i,n}$")
plt.xlabel("Time")
plt.ylabel("Weight")
plt.legend(loc="upper right")
plt.show()

n=15
Bernstein_values=[]
sum_values=np.zeros(1000)
for i in range(0,n+1):
    Bernstein_values.append(Bernstein(i,n,time_values))
    sum_values+=Bernstein_values[i]
    
plt.figure()
plt.title(r"$B_{i,n}(t)$ for $n=15$")
for j in range(n+1):
    plt.plot(time_values,Bernstein_values[j], label=f"B_{j,n}")
plt.plot(time_values, sum_values,color="black", label=r"$\Sigma_{i=0}^15 B_{i,n}$")
plt.xlabel("Time")
plt.ylabel("Weight")
plt.legend(fontsize="5", loc="upper right")
plt.show()
```

In the plots above we have plotted $B_{i,n}(t), \space i=0,\dots,n, \space t \in [0,1]$ for $n=4,5,15$. We have also plotted $\Sigma_{i=0}^n B_{i,n}(t)$ for $t \in [0,1]$, which we observe is contsantly equal to $1$ for all $t$. This confirms the result from $\mathbf{c)}$.

### b) 
Now we evaluate $c(t)$ at its end points $t= 0$ and $t = 1$. Using the results from $\textbf{1b)}$ we obtain 

$$

From the results above we can spesify interpolation properties at it end points. Similarly for the endpoint $t=1$, the first other derivative is determined on the last two points ($p_{n-1}$ and $p_n$) and the order $n$ of the curve. When we change $p_2$ we change the starting acceleration $\vec{a}(0)$ of the curve, but the starting velocity $\vec{v}(0)$ remains the same. ```python
def de_Casteljau(P,t0):
    '''
    function that implements the de Casteljau algorithm

    input:
        P: vector of points we want to make a Bezier Curve for 
        t0: t-parameter 
    output:
        Pvecs[-1,0]: the point on the Bezier curve for t0 
        Pvecs[:,0]: vector of points for every "level" k considered 
    '''
    n,d = np.shape(P) #get dimentions of P 
    Pvecs = np.zeros((n, n, d)) #create matrix of points  
    Pvecs[0,:] = P #initialize the matrix 
    for k in range(1,n): #double for-loop to calculate the Bezier curve at t0 
        for i in range(0,n-k):
            Pvecs[k,i] = (1-t0) * Pvecs[k-1,i] + t0 * Pvecs[k-1, i+1] #de Casteljau's algorithm 
    return Pvecs[-1,0], Pvecs
    

p0 = [0,0]; p1 = [-1,5]; p2 = [4,4]; p3 = [9,3]; p4 = [8,8] #points that we want to make our Bezier curve for 
P = np.array([p0, p1, p2, p3, p4]) #make it into numpy array 


T = np.linspace(0,1, 100) #our time interval 

Pvecs = np.zeros((len(T),  2))#our bezier curve for different k-levels 
Pvecs2 = np.zeros((len(T), len(P),len(P), 2))
for t in range(len(T)): #for loop that uses de Casteljaus algrorithm 
    Pvecs[t] = de_Casteljau(P, T[t])[0]


plt.figure(figsize = (10,7))    
plt.plot(P[:,0], P[:,1], 'r.') #plot the points we want to make Bezier curve for 

plt.plot(Pvecs[:,0], Pvecs[:,1],c='blue', label = f'Bezier curve')

points2 = de_Casteljau(P, 1/3)[1]
plt.plot(points2[0,:,0], points2[0,:,1], 'o', c='r', label='Control points')
plt.plot(points2[0,:,0], points2[0,:,1], 'o--', c='r', label='k=0')
plt.plot(points2[1,:4,0], points2[1,:4,1], 'x--', c='orange', label='k=1')
plt.plot(points2[2,:3,0], points2[2,:3,1], 'x--', c='green', label='k=2')
plt.plot(points2[3,:2,0], points2[3,:2,1], 'x--', c='purple', label='k=3') 
plt.plot(points2[4,:1,0], points2[4,:1,1], 's', c='blue', label=r'$c(\frac{1}{3})$')
plt.grid(True)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Bzier Curve and Intermediate Points')
plt.legend()
plt.show()
```

For points $(p_0,p_1, \ldots, p_n)\in \mathbb{R}^d$ and $(q_0,q_1, \ldots, q_n)\in \mathbb{R}^d$ with their respective Bzier curves $c(t) = b_n(t; p_0,\ldots, p_n)$ and $d(t) = b_n(t; q_0,\ldots, q_n)$, the composite Bzier curve $s(t)$, $s:[0,2] \rightarrow \mathbb{R}^d$ is defined by:

##### 3)
For $s(t)$ to be an $C^1$ function it need to be continious and continious differentiable. We know that $c(t)$ and $d(t)$ seperated are continious and continious differentiable. ```python
def composite_Bezier(P,t):
    '''
    Computes the composite Nezier curve for a set of points P on time interval t 
    input:
        P: points we want to compute the Bezier curve for
        t: time interval points 
    output:
        B: Bezier curve as a function of time, t
    '''
    B = np.zeros((len(t),2)) #array for our Bezier curve 
    i = 0 #indicator of the segment 
    for k in range(len(t)): 
        B[k] = de_Casteljau(P[:,i],t[k]-i)[0] #compute Bezier curve at a discrete point 
        if(int(np.floor(t[k])>i)): #if test that changes the segment 
            i+=1
    return B 
    
P = np.zeros((4, 3, 2)) #the points we want to make Bezier cruve for
P[0,0] = [-1,0];P[1,0] = [-1,1/3]; P[2,0] = [-1/3, 1]; P[3,0] = [0,1]
P[0,1] = [0,1];P[1,1] = [1/3,1]; P[2,1] = [1, 1/3]; P[3,1] = [1,0]
P[0,2] = [1,0];P[1,2] = [1,-1/3]; P[2,2] = [1/3, -1]; P[3,2] = [0,-1]

T = np.linspace(0,3,100) #time array 
B = composite_Bezier(P,T) #Bezier curve 
plt.figure(figsize = (12,10))
plt.title("Bezier curve for test points")
plt.plot(P[:,0,0], P[:,0,1],'.', c="red", label=r'$\mathbf{p}_{j,1}$')
plt.plot(P[:,1,0], P[:,1,1],'.', c="orange", label=r'$\mathbf{p}_{j,2}$')
plt.plot(P[:,2,0], P[:,2,1], '.', c="green", label=r'$\mathbf{p}_{j,3}$')
plt.plot(B[:,0], B[:,1], "blue", label=r'$\mathbf{B}(t)$')
plt.legend(fontsize="14", loc="upper right")
plt.grid()
plt.show()
```

### b) 

We are given the data points $\mathbf{a}_0,\ldots,\mathbf{a}_{m-1} \in \mathbb{R}^d$, and velocities $\mathbf{v}_0, \mathbf{v}_1,\ldots,\mathbf{v}_{m-1} \in \mathbb{R}^d$. We have to find the periodic composite cubic Bzier curve $\mathbf{B}(t)$ such that the following properties holds

* $\mathbf{B}(0) = \mathbf{B}(m)$
* $\mathbf{B}^\prime(0) = \mathbf{B}^\prime(m)$
* $\mathbf{B}(i) = \mathbf{a}_i$ for $i=0,\ldots,m-1$
* $\mathbf{B}\prime(i) = \mathbf{v}_i$ for $i=0,\ldots,m-1$

For $\mathbf{B}(0) = \mathbf{B}(m)$ to hold, we observe that $\mathbf{B}(0)$ the starting point and the end point has to be the same point. Thus
$$\mathbf{B}(0) = \mathbf{p}_{0,1}=\mathbf{a}_{0} = \mathbf{p}_{3,m} = \mathbf{B}(m) \Rightarrow a_0 = \mathbf{B}(0) = \mathbf{B}(m) $$

For $\mathbf{B}^\prime(0) = \mathbf{B}^\prime(m)$ to hold, we observe that the velocity at the start and at the endpoint has to be the same. Thus the result from $2c)$ has to hold.

$$
\begin{cases}
B^{\prime}(0) = 3(p_{1,0}-p_{0,0}) =\mathbf{v}_0  \\
B^{\prime}(m) = 3(p_{3,m} - p_{2,m}) = \mathbf{v}_m=\mathbf{v}_0 \\
\end{cases}
$$

For $\mathbf{B}(i) = \mathbf{a}_i$ we use the original expression for $\mathbf{B}(t)$ and plug in $t=i$

$$
\begin{align}\\
\mathbf{B}(i)&=b_3(1; \mathbf{p}_{0,i}, \mathbf{p}_{1,i},\mathbf{p}_{2,i}, \mathbf{p}_{3,i}) \\ 
&=b_3(0; \mathbf{p}_{0,i+1}, \mathbf{p}_{1,i+1}, \mathbf{p}_{2,i+1}, \mathbf{p}_{3,i+1}) \\ 
&\Rightarrow \mathbf{p}_{3,i}=\mathbf{p}_{0,i+1} = \mathbf{a}_i
\end{align}
 $$
for $i = 1,\ldots,m-1$. For $i=0$ we have already shown that $\mathbf{B}(0) = \mathbf{a}_0$

For $\mathbf{B}\prime(i) = \mathbf{v}_i$ to hold, we observe that the velocity at the start and end of each segment from $i=1$ to $i=m$ has to be the same. Thus the result from $2c)$ can be applied to show the following

$$
\mathbf{v}_{i-1}=\mathbf{B}^\prime(i) = 3(p_{1,i}-p_{0,i})=3(p_{1,i}-a_{i-1}) \\
\Rightarrow p_{1,i}=a_{i-1}+\frac{v_{i-1}}{3}
$$

for $i=1,\ldots,m$

$$
\mathbf{v}_{i}=\mathbf{B}^\prime(i) = 3(p_{3,i}-p_{2,i})= 3(a_{i}-p_{2,i}) \\
\Rightarrow p_{2,i}=a_{i}-\frac{v_{i}}{3}
$$
for $i=1, \ldots, m-1$. 

For $i=m$ we define $p_{2,m}=a_{0}-\frac{v_{0}}{3}$.

Combining the results above, we define:

$$
\begin{align}
\mathbf{B}(t) &= \left\{ b_3(t - i + 1; \mathbf{p}_{0,i}, \mathbf{p}_{1,i}, \mathbf{p}_{2,i}, \mathbf{p}_{3,i}) \right\} \\
&= \left\{ \mathbf{b}_3(t - i + 1; \mathbf{a}_{i-1}, \mathbf{a}_{i-1} + \frac{\mathbf{v}_{i-1}}{3}, \mathbf{a}_{i} - \frac{\mathbf{v}_{i}}{3}, \mathbf{a}_{i}) \right\}
\end{align}
$$

for $i=1,\ldots,m-1$. For $i=m$ we have
$$
\begin{cases}
\mathbf{p}_{0,m}=\mathbf{a}_{m-1} \\
\mathbf{p}_{1,m}=\mathbf{a}_{m-1} + \frac{\mathbf{v}_{m-1}}{3} \\
\mathbf{p}_{2,m}=\mathbf{a}_{0} - \frac{\mathbf{v}_{0}}{3} \\
\mathbf{p}_{3,m}=\mathbf{a}_{0}
\end{cases}
$$

This function is implemented in the code block below.

```python
def interpolate_periodic(A,V):
    ''' 
    Function that uses the properties of Bezier curve to periodically interpolate points and its velocities 
    The functions returns control points for a Bezier curve 
    
    input: 
        A: datapoints we want to interpolate
        V: velocities at the datapoints
    output:
        P: control points for a Bezier curve 
    '''
    m,d = np.shape(A) #get shape of A 
    P = np.zeros((4,m,d)) #cinstruct array that contains our controlpoints
    for i in range(m-1):
        P[0,i] = A[i]
        P[1,i] = A[i] + V[i]/3
        P[2,i] = A[i+1] - V[i+1]/3
        P[3,i] = A[i+1]
    
    P[0,-1] = A[-1]
    P[1,-1] = A[-1] + V[-1]/3
    P[2,-1] = A[0] - V[0]/3
    P[3,-1] = A[0]
        
    return P

```
